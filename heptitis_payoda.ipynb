```python
##convert boolean(True/False) to int(1/0) with map()
def bool_change(X):
    for i in data:
        X[i].replace(False, 0, inplace=True)
        X[i].replace(True, 1, inplace=True)
        
bool_change(data)
```

```python
#### when you remove the outlier, almost 50% of data as have been
Q1 = data.quantile(0.25)
Q3 = data.quantile(0.75)
IQR = Q3 - Q1
data_out = data[~((data < (Q1 - 1.5 * IQR)) |(data > (Q3 + 1.5 * IQR))).any(axis=1)]
data_out.shape
```

```python
data.head(4)
```

```python
data['steroid']=data['steroid'].fillna(data['steroid']).mode()[0]
data['antivirals']=data['antivirals'].fillna(data['antivirals']).mode()[0]
data['fatigue']=data['fatigue'].fillna(data['fatigue']).mode()[0]
data['malaise']=data['malaise'].fillna(data['malaise']).mode()[0]
data['anorexia']=data['anorexia'].fillna(data['anorexia']).mode()[0]
```

```python
data['liver_big']=data['liver_big'].fillna(data['liver_big']).mode()[0]
data['liver_firm']=data['liver_firm'].fillna(data['liver_firm']).mode()[0]
data['spleen_palpable']=data['spleen_palpable'].fillna(data['spleen_palpable']).mode()[0]
data['spiders']=data['spiders'].fillna(data['spiders']).mode()[0]
data['ascites']=data['ascites'].fillna(data['ascites']).mode()[0]
data['varices']=data['varices'].fillna(data['varices']).mode()[0]
```

```python
data
```

```python
data['bilirubin']=data['bilirubin'].fillna(data['bilirubin']).mode()[0]
data['alk_phosphate']=data['alk_phosphate'].fillna(data['varices']).mean()
data['sgot']=data['sgot'].fillna(data['sgot']).mode()[0]
data['albumin']=data['albumin'].fillna(data['albumin']).mean()
data['protime']=data['protime'].fillna(data['protime']).mean()
```

# Data Visualization 

```python
import seaborn as sns
```

```python
### historical representation of dataset which has categorical value and with age to view the frequency distribution 
plt.rcParams["figure.figsize"] = 20,20
data_cat.hist();
```

```python
### historical representation of dataset which has numerical value to view the frequency distribution and curve structure to determine
# skewness of the dataset
plt.rcParams["figure.figsize"] = 20,20
data_num.hist();
```

```python
### determining correlation of various field of columns 
### representation of correlation matrix of continuous attributes
data_num_cor=data_num.corr()
highlight_corr = data_num_cor[data_num_cor>0.85]
f, ax = plt.subplots(figsize=(10, 10))
ax = sns.heatmap(highlight_corr,cmap = 'coolwarm',linewidths=1,linecolor = 'green')
```

```python
### representation of correlation matrix of categorical attributes
data_cat_cor=data_cat.corr()
highlight_corr_cat= data_cat_cor[data_cat_cor>0.9]
f, ax = plt.subplots(figsize=(10, 10))
ax = sns.heatmap(highlight_corr_cat,cmap = 'coolwarm',linewidths=1,linecolor = 'green')
```

```python
### representation of correlation matrix of actual data to find the significance variable in dataset
## here -albumin,alk_phosphate,bilirubin are correlated , in heptitis is diagonised by 
##various parameter test which can have include same procedure with different chemicalreaction results
data_cor=data.corr()
highlight_corr_data = data_cor[data_cor>0.85]
f, ax = plt.subplots(figsize=(10, 10))
ax = sns.heatmap(highlight_corr_data,cmap = 'coolwarm',linewidths=1,linecolor = 'green')
```
